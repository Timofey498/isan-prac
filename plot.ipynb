```python
n, m = [int(i) for i in input().split()]
matrix = [[0] * m for _ in range(n)]
directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
row, col = 0, 0
direction_index = 0
for num in range(1, n*m + 1):
    matrix[row][col] = num
    dx, dy = directions[direction_index]  
    nr = row + dx
    nc = col + dy 
    if nr < 0 or nr >= n or nc < 0 or nc >= m or matrix[nr][nc] != 0:
        direction_index = (direction_index + 1) % 4
        dx, dy = directions[direction_index]
        nr = row + dx
        nc = col + dy
    row, col = nr, nc 
        




for r in range(n):                    
    for c in range(m):
        print(str(matrix[r][c]).ljust(3), end='')
    print()
```

```python
from turtle import *
from math import sin, cos, pi
def heart():
    t = 0
    a = 2*pi
    screensize()
    speed(0)
    hideturtle()
    pencolor('red')
    begin_fill()
    fillcolor('red')
    while t <= a:
        x = 128*sin(t)**3
        y = 8 * (13*cos(t)-5*cos(2*t)-2*cos(3*t)-cos(4*t)-5)
        goto(x,y)
        t += 0.01
    end_fill()
    penup()
    goto(0,-50)
    pendown()
    write('pe, !')
heart()
```
